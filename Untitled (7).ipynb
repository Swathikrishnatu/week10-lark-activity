4. Carry out correlation analysis on the dataset and find the correlation factor
between

```python
fregraph = data.select_dtypes(include=["float"])
fregraph.hist(figsize=(20,15))
plt.show()
```

```python
corrmatrix =data.corr()
plt.subplots(figsize=(15,10))
sns.heatmap(corrmatrix,vmin=-2,vmax=1,annot=True,linewidth=.2,cmap='YlGnBu')
```

No two variables in the dataset ara highly correlated

a. ph and Chloramines

```python
correlation = data['ph'].corr(data['Chloramines'])
print("correlation factor between Ph and Chloramines :",correlation)
 
```

```python
sns.scatterplot(x=data['ph'], y=data['Chloramines'])
plt.title("Ph vs Chloramines")
plt.show()
```

correlation factor between Ph and Chloramines are approximately equal to zero  and weakly negatively correlated .we can conclude that there is no significant correlation between Ph and Chloramines

b. ph and Conductivity

```python
correlation = data['ph'].corr(data['Conductivity'])
print("correlation factor between Ph and Conductivity :",correlation)
 
```
